```python
#we are converting celsius to farenheit and then displaying it
temp = (-6 * 5/9) + 32
print(temp)
```

```python
#we created a function to convert celsius to fahrenheit
def c_to_f(c_temp):
  fahrenheit_temp = (c_temp * (9/5)) + 32
  return fahrenheit_temp
```

```python
#here we are using our funtion to conver 4 degrees celsius to fahrenheit
c_to_f(4)
```

```python
#we are assigning this temp to a variable now
f_temp = c_to_f(4)
print(f_temp)
```
