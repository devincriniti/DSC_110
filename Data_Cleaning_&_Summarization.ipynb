## Data Cleaning

```python
## this is giving us the amount of missing variables for each column
df.isna().sum()
```

```python
## this is giving the percent of patients with the value missing for the specific variables
df.isna().mean() * 100
```

```python
## here we are getting rid of the rows that dont have data for wether the patient has diabetes or not - we are creating a new dataframe (df_clean) for this data removal/cleaning

df_clean = df.dropna(subset=['has_diabetes']).copy()
```

```python
## here we are verifying wether our cleaning worked or not by reruning the percent average command

df_clean.isna().mean() * 100
```

```python
## here we are verifying that we have less rows, since we removed some with the clean
df_clean.shape
```

```python
## here we are assigning the mean systolic blood pressure to : mean_bp , and having python give us the mean using the print function
mean_bp = df_clean['systolic_bp'].mean()
print(mean_bp)
```

```python
## here we are filling any missing blood pressure values in our data with the mean


df_clean['systolic_bp'] = df_clean['systolic_bp'].fillna(mean_bp)
```

### Transform Variables

```python
## here we are converting numerical values to categorical values
## using the cut function, we are basically converting the numerical bmi to weither their obese, under over or normal weight
df['bmi_category'] = pd.cut(df['bmi'],
                            bins=[0, 18.5, 25, 30, 100],
                            labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
```

```python
## here we are checking our work, and can see that we have successfully inserted the categorical data
df['bmi_category'].head()
```

```python
## Here we are given the number of patients in each BMI category
df['bmi_category'].value_counts()
```

###Categorical Variable Transformation

```python
##here we are looking at the first five values for education, noting that it shows us the value in a number, which is less meaningful than if it was transformed into a categorical variable

df['education'].head()
```

```python
## here we are assigning categories/words for each number in the education data set, this is making it more meaningful because college grad means more than "5"

education_map = {
    1: 'Less than 9th grade',
    2: '9-11th grade',
    3: 'High school grad',
    4: 'Some college or AA degree',
    5: 'College Graduate',
    9: 'Do not know',
    'nan': 'Missing'
}
```

```python
## here we are checking are work and making sure 1 brings up the correct category we just set up
education_map[1]
```

```python
# here we created a new field by converting education from the map we created above,
# we then told the program to count the amount of people that fall in each category

df['education_label'] = df['education'].map(education_map)
df['education_label'].value_counts()
```

###**Data Summarization**

```python
#numerical values
# here we are asking the program to figure out the mean of the age, and assigning it to the variable avg_age, and also having it round to the nearest tenth decimal point
# we also use format string literal to combine meaningful words with the rounded average age
avg_age = df['age'].mean()
print(round(avg_age,0))
print(f'The average age in our data is {round(avg_age,0)}')
```

```python
# here we are having the program figure out the standard deviation for age, we then use format string literals to represent the sd meaningfully
std_age = df['age'].std()
print(f'The standard deviation for age in our data is {round(std_age,0)}')
```

```python
# here we are having the program figure out the median for us
df['age'].median()
```

```python
# here we are having the program figure out the 25 and 75 quartile for us

df['age'].quantile([0.25, 0.75])
```

```python
## here we are having the program figure out the most repeated value or "mode" for the age
df['age'].mode()
```

```python
## gives us a comprehensive list of descriptors of the data
df['age'].describe()
```
